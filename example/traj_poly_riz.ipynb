```python
import copy

def setup_wp(wp:np.ndarray, seg_num):
    res = []    # col 0->3: time, x, y, z
    tmp = np.array([0,1,2])
    wp[:,1:] -= tmp
    
    seg_len_idx = wp.shape[0]//seg_num
    for s in range(seg_num):
        if s < seg_num-1:
            tmp = wp[s*seg_len_idx: (s+1)*seg_len_idx+1, :]
        else:
            tmp = wp[s*seg_len_idx:, :]
        tmpt = tmp[:,0]
        tmpt -= tmpt[0]
        res.append(np.c_[tmpt,tmp[:,1:]])
        
    return res


wp = np.zeros((20, 4))
seg_num = 3
wp[:,0] = np.arange(wp.shape[0])*.1
wp[:,1:] = np.arange(wp.shape[0]*3).reshape((-1,3))
for w in setup_wp(wp, seg_num):
    print(w,end='\n\n')
```
